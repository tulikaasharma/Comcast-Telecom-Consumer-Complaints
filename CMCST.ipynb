```python
def bar_chart(feature):
    Closed = csdb[csdb['Closed']==1][feature].value_counts()
    Open = csdb[csdb['Open']==0][feature].value_counts()
    df = pd.DataFrame([Closed,Open])
    df.index = ['Closed','Open']
    df.plot(kind='bar',stacked=True)
```
